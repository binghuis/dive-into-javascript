js 是基于 ecma 标准实现的编程语言，js 是解释型语言，在第一次执行的时候并不是机器码，因此 js 的执行需要有一个宿主环境，
宿主环境主要是提供一个 js 可以运行的环境，主要包括 js 引擎和一些和宿主环境特性相关的 api。

宿主环境是 js 引擎所在的环境，js 在宿主环境执行的时候能使用宿主环境提供的 API 和宿主对象提供的特性和能力，比如浏览器就是我们平时用的最多的宿主环境，
http 接口和 window 对象等都是浏览器这个宿主环境提供的，js 引擎只负责 js 的执行，它本身是没有浏览器这些能力的，这些 WebApi 都是宿主环境提供的。

编译型语言是提前编译执行，也就是说在执行的时候代码已经被编译成计算机可以理解的机器码。

而解释型语言在初次执行的时候往往并没有经过全量编译，而是通过 js 引擎的解释器解释执行，尽管现代 js 引擎都应用了 JIT 技术，
js 是一边编译一边执行的，但是 js 的编译与编译型语言的提前编译依然有很大不同。

js 的编译执行都是通过 js 引擎实现的。
